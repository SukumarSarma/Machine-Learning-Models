```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
df=load_breast_cancer()
df
```

```python
df.keys()
```

```python
newdf=pd.DataFrame(df.data,columns=[df.feature_names])
print('features=',len(newdf.columns))
newdf.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler().fit_transform(newdf)
scale_df=pd.DataFrame(scaler,columns=[df.feature_names])
scale_df.head()
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2).fit(scale_df)
pcanew=pca.transform(scale_df)
pcanew
```

```python
pcanew[:,0].shape,pcanew[:,1].shape
```

```python
plt.figure(figsize=(10,8))
plt.scatter(pcanew[:,0],pcanew[:,1],c=df.target,cmap='viridis',s=50)
plt.xlabel('PCA1')
plt.ylabel('PCA2')
```
